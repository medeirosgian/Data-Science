```python
# Importar bibliotecas
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import dados
treino = pd.read_csv('https://raw.githubusercontent.com/medeirosgian/datasets/main/HousePricing/train.csv', index_col='Id')
teste = pd.read_csv('https://raw.githubusercontent.com/medeirosgian/datasets/main/HousePricing/test.csv', index_col='Id')

treino.head()
```

```python
# LabelEncoding de dados tipo string
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()

# Obter lista de colunas não numéricas
objList = treino.select_dtypes(include = "object").columns

# Criar um novo DataFrame com dados numéricos
treino_lb = treino
teste_lb = teste

# LabelEncoder
for obj in objList:
  treino_lb[obj] = lb.fit_transform(treino[obj].astype(str))
  teste_lb[obj] = lb.fit_transform(teste[obj].astype(str))

# Remover valores indesejados
treino_lb = treino_lb[np.isfinite(treino_lb).all(1)]
teste_lb = treino_lb[np.isfinite(treino_lb).all(1)]

treino_lb.head()
```

```python
# Criar e importar modelo de regressão
model = LinearRegression()

# Separar variáveis dependentes e independentes
X_treino = treino_lb.drop('SalePrice', axis=1)
y_treino = treino_lb['SalePrice']

X_teste = teste_lb.drop('SalePrice', axis=1)
y_teste = teste_lb['SalePrice']
```

```python
# Integrar dados ao modelo
model.fit(X_treino, y_treino);
```

```python
# Predizer valores com o modelo
y_pred = model.predict(X_teste)
```

```python
# Verificar desempenho do modelo
print('R² Score: \t', r2_score(y_teste, y_pred))
```
