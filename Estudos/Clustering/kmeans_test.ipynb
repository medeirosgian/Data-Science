```python
# Import bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# Import dados
df = pd.read_csv('https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/clustering.csv')
df.head()
```

```python
# Criação modelo
from sklearn.cluster import KMeans

# Criar modelos
model1 = KMeans(n_clusters=2)
model2 = KMeans(n_clusters=4)
model3 = KMeans(n_clusters=6)

# Separar dados
x = df[['ApplicantIncome', 'LoanAmount']]

# Padronizar dados
scaler = StandardScaler().fit(x)
x = scaler.transform(x)
```

```python
# Fit modelo de treino
model1.fit(x)
model2.fit(x)
model3.fit(x)

# Predição
y_pred1 = model1.predict(x)
y_pred2 = model2.predict(x)
y_pred3 = model3.predict(x)
```

```python
# Modelo com 2 clusters
plt.scatter('ApplicantIncome', 'LoanAmount', data=df, c=y_pred1, s=40, cmap='viridis')
plt.show()
```

```python
# Modelo com 4 clusters
plt.scatter('ApplicantIncome', 'LoanAmount', data=df, c=y_pred2, s=40, cmap='viridis')
plt.show()
```

```python
# Modelo com 6 clusters
plt.scatter('ApplicantIncome', 'LoanAmount', data=df, c=y_pred3, s=40, cmap='viridis')
plt.show()
```
